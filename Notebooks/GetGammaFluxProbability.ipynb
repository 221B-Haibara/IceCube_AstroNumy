```python
lc_dir = "../Resources/LightCurve/500LCtext_1GeV4week_V1/"
xbins = None
fluxes = {}
for fname in os.listdir(lc_dir):
    sourcename = fname.split("_")[1].split("_")[0]
    df = pd.read_csv(os.path.join(lc_dir, fname), delimiter="\t")
    fluxes[sourcename] = (df.Flux.sum(), df.Flux.values)
    
    if xbins is None:
        xbins = (df.month - 0.5).values.tolist()
        xbins.append(xbins[-1]+1) # add last bin edge
```

```python
def GetTimeProb(alpha_i, I_gamma, norm): #p(t)
    prob = (1.-alpha_i) / (len(xbins)-1) + alpha_i *  I_gamma / norm
    return  prob
```

```python
def GetFluxProb(alpha_i, I_gamma, norm, n_samples = 10000):
    samples = []
    flux = []
    x_min = np.min(xbins) 
    x_max = np.max(xbins)
    time_pdf = GetTimeProb(alpha_i, I_gamma, norm)
    y_max = np.max(time_pdf)
    counter = 0
    while counter < n_samples:
        xrand = np.digitize(np.random.uniform(x_min, x_max), xbins) - 1
        rand_y = np.random.uniform(0, y_max)
        if rand_y < time_pdf[xrand]:
            #samples.append(xrand)
            flux.append(I_gamma[xrand])
            counter += 1
    return flux       
```

```python
#4% case we observe one source in the error circle
#nudist = NuDist()
output = []
ts = []
alpha_0 = []
I_gamma = []
prob_H0 = []
prob_H1 = []
for _ in xrange(300):
    time = np.random.randint(0,len(xbins)-1)
    #print ra,dec
    
    #select one rnd source
    rndFluxIndex = np.random.randint(len(fluxes))
    norm, rndFlux = fluxes.values()[rndFluxIndex]
    
    #read light curve flux at t
    sampledFlux = rndFlux[time]
    #print "time in months", time, "prob", sampledFlux
    #for 11 different alphas get prob and remember best fitting alpha
    prob0 = None
    alphaMax = 0
    probMax = -1
    for alpha in np.linspace(0,1,11):
        y,fluxBins = np.histogram(GetFluxProb(alpha, rndFlux, norm),normed=True)
        #meow fix me
        #print " meow"
        #print alpha, len(y), len(fluxBins), sampledFlux, np.digitize(sampledFlux,fluxBins), fluxBins
        #print "end meow"
        binx = np.digitize(sampledFlux,fluxBins)-1
        if binx >= len(y) or binx < 0:
            prob = 0
        else:
            prob = y[binx]
        if prob0 is None:
            prob0 = prob
        if prob>probMax:
            probMax = prob
            alphaMax = alpha
        #print alpha, np.log(prob)
        output.append(np.log(prob))
    if prob0==0:
        ts.append(1.)
        prob_H0.append(0.)
        prob_H1.append(0.)
    else:
        ts.append(probMax/prob0)
        prob_H0.append(prob0)
        prob_H1.append(probMax)
    #print rndFluxIndex, time
    #print prob, prob0, probMax, alphaMax
    alpha_0.append(alphaMax)
    I_gamma.append(sampledFlux)
```
