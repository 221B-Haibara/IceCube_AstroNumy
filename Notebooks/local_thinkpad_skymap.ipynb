```python
%%time
angular_res = 0.5
within_res_1 = []
for ra,dec in nudist.get_random(100000):
    delta_ra = fermi.ra - ra
    delta_dec = fermi.dec - dec
    r = sqrt(delta_ra**2 + delta_dec**2)
    sel = r < angular_res
    n = sel.sum()
    within_res_1.append(n)
    
    
```

```python
a = plt.hist(within_res_1, bins=np.linspace(-0.5,9.5,11), normed=True)
plt.yscale("log",nonposy='clip')
plt.ylim(1e-5,1e1)
plt.xlabel("number of sources in the error circle")
plt.ylabel("counts")
```
